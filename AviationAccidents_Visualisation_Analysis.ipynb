```python
countries = ['Canada', 'Brazil', 'United States', 'United Kingdom', 'Spain']
fill_colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854']
gf = df.groupby('Country')
```

```python
datar = []
for country, fill_color in zip(countries[::-1], fill_colors):
    group = gf.get_group(country)
    years = (group['Event.Date'].dt.year).tolist()
    length = len(years)
    country_coords = [country] * length
    pop = group['Total.Fatal.Injuries'].tolist()
    zeros = [0] * length
    datar.append(dict(
        type='scatter3d',
        mode='lines',
        x=years + years[::-1] + [years[0]],  # year loop: in incr. order then in decr. order then years[0]
        y=country_coords * 2 + [country_coords[0]],
        z=pop + zeros + [pop[0]],
        name='',
        surfaceaxis=1, # add a surface axis ('1' refers to axes[1] i.e. the y-axis)
        surfacecolor=fill_color,
        line=dict(
            color='black',
            width=4
        ),
    ))
```

```python
tdlayout = dict(
    title='Aviation Fatalities',
    showlegend=False,
    scene=dict(
        xaxis=dict(title=''),
        yaxis=dict(title=''),
        zaxis=dict(title=''),
        camera=dict(
            eye=dict(x=-1.7, y=-1.7, z=0.5)
        )
    )
)  
```

```python
fig = dict(data=datar, layout=tdlayout)
```
